```python
df_str = df2[["Issue Date", "Vehicle Body Type", "Vehicle Make", "Violation Time", "Vehicle Color"]]
df_str
```

```python
df_str = df_str.rename(columns={'Issue Date': 'Date','Vehicle Body Type': 'Type', 'Vehicle Make': 'Make', 'Violation Time':'Time', 'Vehicle Color': 'Color'})
```

```python
df_str
```

```python
types_df = df_str.groupby(['Type']).size()
types_df = types_df.sort_values(ascending=False)
types_df
```

```python
types_counts = df_str.groupby(['Type']).size().reset_index(name='counts')
types_counts
```

```python
makes_df=df_str.groupby(['Make']).size()
makes_df = makes_df.sort_values(ascending=False)
makes_df
```

```python
colors_df = df_str.groupby(['Color']).size()
colors_df = colors_df.sort_values(ascending=False)
colors_df
```

```python
colors_count = df_str.groupby(['Color']).size().reset_index(name='counts')
colors_count
```

```python
types_df.to_csv(typescount)
makes_df.to_csv(makescount)
colors_df.to_csv(colorscount)
```
